```python
#Response time with our model with tamil test dateset
import pandas as pd
from transformers import pipeline, AutoModelForQuestionAnswering, AutoTokenizer
import time
```

```python
# === Load model ===
model_name = "RajeevanL/tamil-distilled-roberta-v3"  # Replace with any model you want
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)
```

```python
# === Load test dataset ===
df = pd.read_csv("/content/test.csv")
```

```python
# === Measure response times ===
times = []

for _, row in df.iterrows():
    qa_input = {"question": row["Question"], "context": row["Context"]}
    start = time.time()
    try:
        _ = qa_pipeline(qa_input)["answer"]
    except Exception:
        pass
    end = time.time()
    times.append(end - start)

# === Average time ===
avg_time = sum(times) / len(times)
print(f"Average response time for model '{model_name}' on test.csv: {avg_time:.4f} seconds")
```
